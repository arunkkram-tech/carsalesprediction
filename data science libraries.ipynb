```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline  # in line ensure the display in notebook result , rather than storoing the display in a file

dataframe=pd.read_csv('C:/Users/Acer/12JAN-18JAN2025-GENAI/DATASETS/Car details v3.csv')
#print(dataframe.head())
```

```python
#GET A RECOMMENDED PRICE OF A SECOND HAND CAR 
```

```python
car_df=dataframe

# info function on the data frame gives the general and structural information on the data structure
#car_df.info()

car_df['mileage']=car_df['mileage'].str.replace('kmpl','')
car_df['mileage']=pd.to_numeric(car_df['mileage'],errors='coerce')


car_df['max_power']=car_df['max_power'].str.replace('bhp','')
car_df['max_power']=pd.to_numeric(car_df['max_power'],errors='coerce')


#\d regular expression for extracing digits#
car_df['torque']=car_df['torque'].str.extract(r'(\d+(rpm))')[0]
car_df['torque']=car_df['torque'].str.replace('rpm','')
car_df['torque']=pd.to_numeric(car_df['torque'],errors='coerce')

car_df.head(5)

#SPLIT THE DATAFRAME INTO TWO DEPEND ON THE DATA TYPPE
numcols=car_df.select_dtypes(include=np.number)
objcols=car_df.select_dtypes(include=['object'])

numcols.describe()

```

```python
#the describe function gives the count , mean , standard deviation etc. It can be used as a baseline for imputation/missing data
#calculation


#Missing values must be imputed./ filled
#              - Manual imputation - by mean/medial imputation

```

```python
#the given loop fills the missing values with median
for col in numcols.columns:
    numcols[col]=numcols[col].fillna(numcols[col].median())

numcols.describe()
```

```python
#drop or convert a column type
#delta - difference between two dates
numcols['age']=2024-numcols['year']


numcols.describe()
```

```python
#drop these columns are from the dataset , if it is unwanted
#AXIS =0 ROWS  , AXIS=1 columns
numcols=numcols.drop(['year','selling_price'],axis=1)
```

```python
#scaling must be done on important column if its values use different  units. THus we should bring uniformity

#SCALERS
# Eg.  Standard Scalar , Min Max Scalar and Robut Scalar
#. 1 STD SCALAR - scales data into both +VE and -VE values with in a range.  
#     (X-mean)/Std Deviation

#. 2 MIN  MAX SCALAR scales data into a range of 0 to 1 and positive only
#     (X-Xmininum)/(XMaximum-Xminimum)

#. 3 ROBUST SCALAR scales data into -VE and +VE with in a range
#     (X-Q2)/(Q3-Q1)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler

numcols_standard_scale=StandardScaler().fit_transform(numcols)

numcols_standard_scale=pd.DataFrame(numcols_standard_scale,columns=numcols.columns)

numcols_standard_scale.head(5)

```

```python
numcols_minmax=MinMaxScaler().fit_transform(numcols)

numcols_minmax=pd.DataFrame(numcols_minmax,columns=numcols.columns)

numcols_minmax.head(5)
```

```python
numcols_robust=RobustScaler().fit_transform(numcols)

numcols_robust=pd.DataFrame(numcols_robust,columns=numcols.columns)

numcols_robust.head(5)
```
